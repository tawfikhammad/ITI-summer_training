# Problem 1

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
# prompt: Use NumPy to create simulations and compute empirical probability for the following outcomes. Compare the results with the theoretical probability.
# Two fair coin flips produce exactly two heads

# Simulate 1 million pairs of coin flips
flips = np.random.randint(0, 2, size=(1000000, 2))

# Count the number of pairs that sum to 2 (both heads)
num_two_heads = np.sum(np.sum(flips, axis=1) == 2)

# Calculate the empirical probability
empirical_prob = num_two_heads / 1000000

# Theoretical probability
theoretical_prob = 0.25  # (1/2) * (1/2)

print("Empirical Probability:", empirical_prob)
print("Theoretical Probability:", theoretical_prob)
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

flips = np.random.randint(0, 2, size=(1000, 3))
num_one_heads = np.sum(np.sum(flips, axis=1) == 1)

print("Empirical Probability:", num_one_heads/1000)
print("Theoretical Probability:", 0.375)
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
#type your code here:

import numpy as np

results = np.random.binomial(3, 0.6, 10000000)

probability = np.mean(results == 1)

print(f"Empirical probability  {probability:.4f}")
```

# Problem 2

The probability of rain in a day is p = 0.5. Give the python code to simulate the probability of rain during weekend (**i.e rain on any one of the 2 weekend days Friday Or Saturday**) (assume independence between raining days). Compare the results with the theoretical probability.

```python
#type your code here

rain_friday = np.random.binomial(1, 0.5, 100000)
rain_saturday = np.random.binomial(1, 0.5, 100000)

# Check if it rains on at least one of the two days
rain_weekend = (rain_friday == 1) | (rain_saturday == 1)

Empirical_probability = np.mean(rain_weekend)

theoretical_probability = 1 - (1 - 0.5) ** 2

print(f"Empirical Probability: {Empirical_probability:.4f}")
print(f"Theoretical Probability: {theoretical_probability:.4f}")
```

# Problem 3
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
#type your code here:

rolls = np.random.randint(1, 7, 1000)

sample_mean = np.mean(rolls)
sample_variance = np.var(rolls, ddof=1)
sample_std_dev = np.std(rolls, ddof=1)

theoretical_mean = (7) / 2
theoretical_variance = (6**2 - 1) / 12
theoretical_std_dev = np.sqrt(theoretical_variance)

# Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Sample Standard Deviation: {sample_std_dev:.4f}")

print(f"Theoretical Mean: {theoretical_mean:.4f}")
print(f"Theoretical Variance: {theoretical_variance:.4f}")
print(f"Theoretical Standard Deviation: {theoretical_std_dev:.4f}")
```
